```python
from flask import Flask, request, abort
from linebot import LineBotApi, WebhookHandler
from linebot.exceptions import InvalidSignatureError
from linebot.models import *
import urllib.request
import json
import os
import ssl
 
def allowSelfSignedHttps(allowed):
    # bypass the server certificate verification on client side
    if allowed and not os.environ.get('PYTHONHTTPSVERIFY', '') and getattr(ssl, '_create_unverified_context', None):
        ssl._create_default_https_context = ssl._create_unverified_context
 
allowSelfSignedHttps(True) # this line is needed if you use self-signed certificate in your scoring service.
 
 
def model_predict(academic1,academic2,depression1,depression2,depression3,depression4,depression5,depression6,sex):
    # Request data goes here
    data = {
        "Inputs": {
            "input1":
            [
                {
                    "id1": 1,
                    "academic1": academic1,
                    "academic2": academic2,
                    "depression1": depression1,
                    "depression2": depression2,
                    "depression3": depression3,
                    "depression4": depression4,
                    "depression5": depression5,
                    "depression6": depression6,
                    "sex": sex
                },
            ]
        },
        "GlobalParameters": {
        }
    }
 
    body = str.encode(json.dumps(data))
 
    url = 'http://b022a4ce-34f7-4c3c-a673-a7aaa17bfec7.eastus2.azurecontainer.io/score'
    api_key = 'gZt2e963bey9j9lS9LGs1KDYVC7EnQcd' # Replace this with the API key for the web service
    headers = {'Content-Type':'application/json', 'Authorization':('Bearer '+ api_key)}
 
    req = urllib.request.Request(url, body, headers)
 
    try:
        response = urllib.request.urlopen(req)
        result = response.read()
        result_decode = result.decode('utf-8')
        return json.loads(result_decode)
    except urllib.error.HTTPError as error:
        print("The request failed with status code: " + str(error.code))
        # Print the headers - they include the requert ID and the timestamp, which are useful for debugging the failure
        print(error.info())
        print(error.read().decode("utf8", 'ignore'))
app = Flask(__name__)

# 填入自己的line bot 金鑰與密碼
channel_access_token = ''
channel_secret = ''
line_bot_api = LineBotApi(channel_access_token)
handler = WebhookHandler(channel_secret)


@app.route("/callback", methods=['POST'])
def callback():
    signature = request.headers['X-Line-Signature']
    body = request.get_data(as_text=True)
    app.logger.info("Request body: " + body)
    try:
        handler.handle(body, signature)
    except InvalidSignatureError:
        abort(400)
    return 'OK'

# 儲存使用者輸入的多筆資料
user_inputs = []

@handler.add(MessageEvent, message=TextMessage)
def handle_message(event):
    user_input = event.message.text  # 使用者輸入的訊息文字
    user_inputs.append(user_input)  # 儲存使用者輸入的資料

    # 判斷是否已經收集到足夠的資料
    if len(user_inputs) >= 9:
        # 將多筆資料作為參數傳遞到 model_predict 函式中進行預測
        result = model_predict(user_inputs[0], user_inputs[1], user_inputs[2], user_inputs[3], user_inputs[4],
                               user_inputs[5], user_inputs[6], user_inputs[7], user_inputs[8])

        # 取得預測結果
        prediction = round(float(result['Results']['WebServiceOutput0'][0]['Scored Labels']), 2)

        # 根據預測結果進行下一步處理
        if prediction < 2.5 and (int(user_inputs[7]) > 3) or prediction > 2.5:
            risk_level = "高危險族群。建議立即使用協助資源(包括諮商、就診)。\n衛福部心理健康司：\nhttps://dep.mohw.gov.tw/domhaoh/mp-107.html"
        elif 2 < prediction < 2.5 or prediction < 2 and 3<= int(user_inputs[7]) < 4:
            risk_level = "中危險族群。\n建議多觀察狀況適時提供關心，若情況惡化請盡速通報校園輔導室"
        elif prediction < 2 or 0 < int(user_inputs[7]) <= 2:
            risk_level = "低危險族群。\n恭喜你！請繼續保持現在的狀態！"
        else:
            risk_level = "無法判斷，請重新開始"

        # 回覆使用者預測結果和風險等級
        reply_text = f"預測結果：{prediction}\n風險等級：{risk_level}"
        end_message = "\n本次測驗到此結束，感謝您的使用！"
        additional_message = "您好，歡迎參加此次心理壓力狀態測驗！\n請依據以下提示進行作答！\n第1題：以這學期來說，你覺得你的成績在班上的排名落在哪個範圍？(1=第1-5名；2=第6-10名；3=第11-15名；4=第16-20名；5=第21名以後)？\n第2題：你覺得功課會給你帶來壓力嗎？(1=沒什麼壓力；2=中等程度；3=壓力蠻大的)？\n第3題：在「過去兩週」你有多常感覺孤單？ \n第4題：在「過去兩週」你有多常感覺憂鬱或悲傷？ \n第5題：在「過去兩週」你有多常感覺對事物不感興趣？\n第6題：在「過去兩週」你有多常對未來感到絕望？\n第7題：在「過去兩週」你多常感到自己無價值？(1從未~5頻繁)\n第8題：在「過去兩週」你多常有想結束生命念頭？ \n第9題：性別(男性填1、女性填2)？\n第3題~第8題依據感受到情緒程度填入以下數值： 1=從未；2=幾乎沒有；3=有時候；4=時常；5=非常頻繁"
        # 先發送回覆訊息
        reply_text = reply_text + end_message
        reply_message = TextSendMessage(text=reply_text)
        line_bot_api.reply_message(event.reply_token, reply_message)

        # 再發送額外訊息
        additional_message = TextSendMessage(text=additional_message)
        line_bot_api.push_message(event.source.user_id, additional_message)



        # 清空使用者輸入的資料
        user_inputs.clear()
    else:
        # 提示使用者繼續輸入資料
        remaining_inputs = 9 - len(user_inputs)
        if len(user_inputs) == 0:
            start_message = "現在開始測驗，請依據提示輸入數字。"
            reply_text = start_message + f"\n請輸入剩餘的 {remaining_inputs} 筆資料"
        else:
            reply_text = f"請輸入剩餘的 {remaining_inputs} 筆資料"
        message = TextSendMessage(text=reply_text)
        line_bot_api.reply_message(event.reply_token, message)



# @handler.add(MessageEvent, message=TextMessage)
# def handle_message(event):
#     # data_arr=[academic1,academic2,depression1,depression2,depression3,depression4,depression5,depression6,sex]
#     # for i in data_arr:
#     message="academic1"
#     line_bot_api.reply_message(event.reply_token, message)
#     # result = model_predict(form_data['age'],form_data['sex'],form_data['bmi'],form_data['children'],form_data['smoker'],form_data['region'])
#     # message = result['Results']['WebServiceOutput0'][0]['Scored Labels']
#     # line_bot_api.reply_message(event.reply_token, message)

import os
if __name__ == "__main__":
    port = int(os.environ.get('PORT', 5000))
    app.run(host='127.0.0.1', port=port)
```
