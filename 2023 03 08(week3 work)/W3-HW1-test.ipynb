#### 第二題
    問：各縣市的學生視力不良率是否呈現逐年攀升？

計算不同年度各縣市成長率

```python
import ezodf

# 讀取 ods 檔案
doc = ezodf.opendoc('W3 homework.ods')

# 選擇要分析的工作表
sheet = doc.sheets['國小視力不良率-縣市']

# 讀取縣市名稱和學年度
cities = [cell.value for cell in sheet.column(0)[1:]]
years = [cell.value for cell in sheet.row(0)[1:]]

# 讀取不良率數據
data = []
growth_rates = []  # 添加一个空的列表以确保与城市相对应的子列表存在
for row in list(sheet.rows())[1:]:
    data_row = []
    for cell in row[1:]:
        if cell.value is not None:
            data_row.append(float(cell.value))
        else:
            data_row.append(None)
    data.append(data_row)
    growth_rates.append([])  # 添加一个空的子列表

# 計算成長率
for i, city_data in enumerate(data):
    for j in range(len(years) - 1):
        if city_data[j] is not None and city_data[j+1] is not None:
            growth_rate = (city_data[j+1] - city_data[j]) / city_data[j] * 100
            growth_rates[i].append(growth_rate)
        else:
            growth_rates[i].append(None)
    # 確保 growth_rates[i] 至少有 len(years) - 1 個元素
    while len(growth_rates[i]) < len(years) - 1:
        growth_rates[i].append(None)

# 列出結果
print("縣市名稱\t\t", end="")
for year in years[1:]:
    print(f"{year}\t", end="")
print()
print("="*70)

for i, city in enumerate(cities):
    print(f"{city}\t", end="")
    if len(growth_rates[i]) < len(years) - 1:
        growth_rates[i] += [None] * (len(years) - 1 - len(growth_rates[i]))
    for j, year in enumerate(years):
        if j == 0:
            continue
        if growth_rates[i][j-1] is not None:
            print(f"{growth_rates[i][j-1]:.2f}%\t", end="")
        else:
            print("N/A\t", end="")
    print()
```

```python
import ezodf
import matplotlib.pyplot as plt

# 讀取 ods 檔案
doc = ezodf.opendoc('W3 homework.ods')

# 選擇要分析的工作表
sheet = doc.sheets['國小視力不良率-縣市']

# 讀取縣市名稱和學年度
cities = [cell.value for cell in sheet.column(0)[1:]]
years = [cell.value for cell in sheet.row(0)[1:]]

# 讀取不良率數據
data = []
growth_rates = []  # 添加一个空的列表以确保与城市相对应的子列表存在
for row in list(sheet.rows())[1:]:
    data_row = []
    for cell in row[1:]:
        if cell.value is not None:
            data_row.append(float(cell.value))
        else:
            data_row.append(None)
    data.append(data_row)
    growth_rates.append([])  # 添加一个空的子列表

# 計算成長率
for i, city_data in enumerate(data):
    for j in range(len(years) - 1):
        if city_data[j] is not None and city_data[j+1] is not None:
            growth_rate = (city_data[j+1] - city_data[j]) / city_data[j] * 100
            growth_rates[i].append(growth_rate)
        else:
            growth_rates[i].append(None)
    # 確保 growth_rates[i] 至少有 len(years) - 1 個元素
    while len(growth_rates[i]) < len(years) - 1:
        growth_rates[i].append(None)

# 繪製成長率圖表
fig, ax = plt.subplots(figsize=(12, 8))
ax.set_title('國小視力不良率成長率', fontsize=20)
ax.set_xlabel('學年度', fontsize=16)
ax.set_ylabel('成長率 (%)', fontsize=16)
for i, city in enumerate(cities):
    if len(growth_rates[i]) < len(years) - 1:
        growth_rates[i] += [None] * (len(years) - 1 - len(growth_rates[i]))
    ax.plot(years[1:], growth_rates[i], label=city)
ax.legend(fontsize=12)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'SimHei'

cities = ['北京市', '上海市', '广州市', '深圳市']
years = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]

data = {'北京市': [1.96, 2.05, 1.99, 1.93, 1.87, 1.80, 1.74, 1.67, 1.61],
        '上海市': [2.16, 2.19, 2.18, 2.16, 2.13, 2.08, 2.04, 1.98, 1.93],
        '广州市': [2.87, 2.87, 2.83, 2.79, 2.72, 2.66, 2.58, 2.50, 2.42, 2.35],
        '深圳市': [4.39, 4.24, 4.09, 3.92, 3.74, 3.53, 3.35, 3.17, 3.01, 2.84]}

growth_rates = []
for city in cities:
    rates = []
    for i in range(len(data[city]) - 1):
        rate = (data[city][i+1] - data[city][i]) / data[city][i] * 100
        rates.append(round(rate, 2))
    growth_rates.append(rates)

fig, ax = plt.subplots()
for i, city in enumerate(cities):
    if len(growth_rates[i]) < len(years) - 1:
        growth_rates[i] += [None] * (len(years) - 1 - len(growth_rates[i]))
    ax.plot(years[1:], growth_rates[i], label=city)
ax.legend(fontsize=12)
plt.show()
```
