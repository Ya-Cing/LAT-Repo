```python
import pandas as pd

df = pd.read_csv('test2 2015.csv')

df.head()
```

```python
temp = df[['Complex4Name', 'FValue']]
temp
```

```python
Type = temp.groupby(['Complex4Name'])['FValue'].sum().reset_index()
```

```python
Type['year'] = 2015
Type
```

```python
df2016 = pd.read_csv('test2 2016.csv')
df2017 = pd.read_csv('test2 2017.csv')
```

```python
new = df2016[['Complex4Name', 'FValue']].groupby(['Complex4Name'])['FValue'].sum().reset_index()
new['year'] = 2016
new
```

```python
new = df2017[['Complex4Name', 'FValue']].groupby(['Complex4Name'])['FValue'].sum().reset_index()
new['year'] = 2017
new
```

```python
frames = [Type, new]

result = pd.concat(frames)
result
```
